**if-elif-else**

- when there is one condition is there
  - in that first answer under **if** block
  - second answer under **else** block

- when there are two condition 
  - we can expect three answer
    - in that first answer under **if** block based on condition-1
    - second answer under **elif** block based on condition-2
    - third answer under **else** block

```python
one condition ===== two answer ===== if-else
two condition ===== 3 answer ====== if-else
```

```python
if <condition>:
    <write your code>
elif<condition>:
    <write your code>
else:
    <write your code>
```

```python
#wap ask the user enter a number
#if the number equal to 1 then print one
#otherwise if number equal to 2 then print two 
#otherwise if the number equal to 3 then print three 
#other wise print enter a valid number

num=eval(input('enter a number :'))
if num==1:
    print('one')
elif num==2:
    print('two')
elif num==3:
    print('three')
else:
    print('enter a valid number')
```

```python
# wap ask a user enter percentage 
# if percentage is greater than 75 print A grade
# elif percentage btweeen 60 to 75 print B grade 
# elif percentage between 45 to 60 printt c grade
# elif percentage between 30 to 45 print D grade
# otherwise print fail

marks=eval(input('enter percentage :'))
if marks >75:
    print('A grade')
elif marks >=60:
    print('B grade')
elif marks >=45:
    print('C grade')
elif marks >=30:
    print('D grade')
else:
    print('fail')
```

**try-except**


- try except is used to handle the exception mechanism

- try and except is used to catch and handle the errors

- when we get the errors python will stop the execution

- using try exception eventhough we get the error, the execution will not stop

- it will catch all the logical errors automatically

- the actual code block will write in **try block**

- if any error occures in try block, we will catch the errors in **except block**

- syntax errors will not catch

```python
# wap user salary 
# ask the user emter the tax_per
# cal how much need to pay

salary=eval(input('enter the salary'))
tax_per=eval(input('enter the tax_per'))
tax=salary*tax_per/100
print(f'total tax is:{tax}')
```

```python
try:
    salary=eval(input('enter the salary'))
    tax_per=eval(input('enter the tax_per'))
    tax=salary*tax_per/100
    print(f'total tax is:{tax}')
except Exception as e:
    print(e) 
```

```python
#even-odd

try:
    num=eval(input('enter the  number'))
    if num%2==0:
        print(f'{num}is even num')
    else:
        print(f'{num}odd num')
except Exception as e:
    print(e)
    
```

```python
**try-except-finally**

- originally code in **try** block

- error catch in **except** block

- **finally** block will activate either or try and except activated

- assume that try block excecuted then finally block can also execute

- assume that except block executed then also finally block executed

- coffee well=====try clean the kitchen ====finally

- co
```

```python
try:
    print('connection established')
except Exception as e:
    print(e)
finally:
    print('connection closed')
```
